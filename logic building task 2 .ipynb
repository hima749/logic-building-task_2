```python
#Problem 1: Unique Words in a Sentence
sentence = input("enter sentence:")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
#Problem 2: Highest Salary from Employee Data

employees = {
    "Rahul": 50000,
    "Anita": 65000,
    "Vikram": 72000,
    "Neha": 68000
}
highest_salary_employee = max(employees, key=employees.get)
print("Employee with highest salary:", highest_salary_employee)
print("Highest salary:", employees[highest_salary_employee])
```

```python
#Problem 3: Find Maximum and Minimum Values
numbers = [12, 45, 7, 89, 23, 50]
maximum = numbers[0]
minimum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num
print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = [800, 1400, 1500, 600, 900, 500]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("product above 1000:", count)
```

```python
#Problem 5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
present_count = 0
for day in attendance:
    if day == "P":
        present_count = present_count + 1
total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100
print("Attendance Percentage:", attendance_percentage)
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

```python
#Problem 7: Count Character Frequency
text = "python"
char_count = {}
for ch in text:
    if ch in char_count:
        char_count[ch] = char_count[ch] + 1
    else:
        char_count[ch] = 1
print(char_count)

```

```python
#Problem 9: Check if a Key Exists in a Dictionary
employees = {
    "emp1": "Rahul",
    "emp2": "Anita",
    "emp3": "Vikram"
}
key_to_check = "emp2"
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#Problem 10: Calculate Average Marks
marks = [70, 75, 80, 65]
total = 0
for m in marks:
    total = total + m
average = total / len(marks)
print("Average Marks:", average)


```
